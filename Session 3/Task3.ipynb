```python
# 1. Load IPL Dataset and Count Missing Values
import pandas as pd
from sklearn.preprocessing import LabelEncoder


# Load dataset
df = pd.read_csv("ipl_sample_data.csv")

# Display first 5 rows
print(df.head())

# Count missing values in each column
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

```python
# 2. Fill Missing Values in 'player_age' with Median
# Find median age
median_age = df['player_age'].median()

# Fill missing values
df['player_age'] = df['player_age'].fillna(median_age)

# Display updated column
print(df['player_age'])
```

```python
# 3. Replace Missing Team Names with 'Unknown'
# Replace missing team names
df['team'] = df['team'].fillna('Unknown')

# Print first 10 rows
print(df.head(10))
```

```python
# 4. One-Hot Encoding on 'venue' Column
# Before encoding

# One-Hot Encoding
venue_encoded = pd.get_dummies(df, columns=['venue'])

print(venue_encoded.head())
```

```python
# 5. Label Encoding on 'player_role'

# Create encoder
le = LabelEncoder()

# Encode player_role
df['player_role_encoded'] = le.fit_transform(df['player_role'])

# Display DataFrame
print(df[['player_role', 'player_role_encoded']])

# Display mapping
print("\nRole Mapping:")

for role, value in zip(le.classes_, range(len(le.classes_))):
    print(f"{role} --> {value}")
```
