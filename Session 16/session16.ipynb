TP = 50
TN = 120
FP = 10

Accuracy = (TP + TN) / (TP + TN + FP + FN)

Step 2: Precision
Precision = TP / (TP + FP)

Precision = 50 / (50 + 10)

Precision = 50 / 60

Precision = 0.8333
Answer
Precision = 83.33%
Step 3: Recall
Recall = TP / (TP + FN)

Recall = 50 / (50 + 20)

Recall = 50 / 70

Recall = 0.7143
Answer
Recall = 71.43%
Step 4: F1 Score
F1 = 2 × (Precision × Recall)/(Precision + Recall)

F1 = 2 × (0.8333 × 0.7143)/(0.8333 + 0.7143)

F1 = 0.7692
Answer
F1 Score = 76.92%

```python
#Task 2: Python Function to Print Confusion Matrix

def print_confusion_matrix(tp, tn, fp, fn):

    print("\nCONFUSION MATRIX")
    print("-" * 35)

    print(f"{'':15} Pred Positive  Pred Negative")

    print(f"Actual Positive     {tp:^10}      {fn:^10}")

    print(f"Actual Negative     {fp:^10}      {tn:^10}")

    print("-" * 35)


# Example
print_confusion_matrix(
    tp=50,
    tn=120,
    fp=10,
    fn=20
)
```

```python
tp = 60
fp = 10
fn = 30

precision = tp / (tp + fp)

recall = tp / (tp + fn)

print("Precision:", round(precision,4))
print("Recall:", round(recall,4))
```

Task 4: Food Delivery Fraud Detection System
Given
TP = 40
TN = 150
FP = 20
FN = 10

```python
tp = 40
fp = 20
fn = 10

# Precision
precision = tp / (tp + fp)

# Recall
recall = tp / (tp + fn)

# F1 Score
f1_score = 2 * (precision * recall) / (precision + recall)

print("Precision:", round(precision,4))
print("Recall:", round(recall,4))
print("F1 Score:", round(f1_score,4))
```
