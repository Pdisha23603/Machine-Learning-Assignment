```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv("mobile_phone_prices.csv")
print(df.head())
```

```python
# Features and Target
X_initial = df[['RAM_GB', 'Storage_GB', 'Battery_mAh', 'Camera_MP']]
y = df['Price_USD']

# Train Model
model_1 = LinearRegression()
model_1.fit(X_initial, y)

print("=== INITIAL MODEL PARAMETERS ===")
print(f"Intercept: {model_1.intercept_:.2f}")

for col, coef in zip(X_initial.columns, model_1.coef_):
    print(f"{col}: {coef:.2f}")

y_pred = model_1.predict(X_initial)

print(f"R² Score: {r2_score(y, y_pred):.4f}")
```

```python
# VIF Calculation
def calculate_vif(df_features):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = df_features.columns
    vif_data["VIF"] = [
        variance_inflation_factor(df_features.values, i)
        for i in range(df_features.shape[1])
    ]
    return vif_data

print("\n=== VIF VALUES ===")
print(calculate_vif(X_initial))
```

```python
# Refactored Model
X_refactored = df[['RAM_GB', 'Battery_mAh', 'Camera_MP']]

model_2 = LinearRegression()
model_2.fit(X_refactored, y)

print("\n=== REFACTORED MODEL ===")

for col, coef in zip(X_refactored.columns, model_2.coef_):
    print(f"{col}: {coef:.2f}")

print(
    f"R² Score: "
    f"{r2_score(y, model_2.predict(X_refactored)):.4f}"
)
```

```python
# Save Updated Dataset
df.to_csv("mobile_price_data_updated.csv", index=False)
print("File Saved Successfully")
```
