```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# TASK 1: Generate Synthetic Data & Base Plot
# Seeding for reproducibility
np.random.seed(42)

# Generate data for a week (7 days)
# X: Instagram hours (ranging roughly from 1 to 5 hours)
instagram_hours = np.array([1.5, 3.0, 4.5, 2.0, 5.0, 1.0, 3.5]).reshape(-1, 1)

# Y: Battery % (generally decreases as Instagram hours increase, with some random noise)
# Formula used implicitly: battery = 80 - 10 * hours + noise
battery_percentage = np.array([68, 48, 32, 62, 25, 76, 40])

# Initialize the plot
plt.figure(figsize=(8, 6))
plt.scatter(instagram_hours, battery_percentage, color='blue', label='Actual Data', s=100)
```

```python
# TASK 2: Fit Linear Regression Model
model = LinearRegression()
model.fit(instagram_hours, battery_percentage)

# Extract slope (m) and intercept (c)
slope = model.coef_[0]
intercept = model.intercept_

print("--- Model Parameters ---")
print(f"Slope (Coefficient): {slope:.2f}")
print(f"Intercept: {intercept:.2f}\n")
```

```python
# TASK 3: Predict and Plot Regression Line
# Generate predictions
predicted_battery = model.predict(instagram_hours)

# Plot the regression line
plt.plot(instagram_hours, predicted_battery, color='red', linewidth=2, label=f'Regression Line (y = {slope:.2f}x + {intercept:.2f})')

# Customizing the plot
plt.title('Instagram Usage vs. End-of-Day Battery Percentage', fontsize=14)
plt.xlabel('Hours Spent on Instagram', fontsize=12)
plt.ylabel('Battery Percentage (%)', fontsize=12)
plt.xlim(0, 6)
plt.ylim(0, 100)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Display the plot
plt.show()
```

```python
# TASK 4: Calculate Mean Squared Error (MSE)
mse = mean_squared_error(battery_percentage, predicted_battery)

print("--- Model Evaluation ---")
print(f"Mean Squared Error (MSE): {mse:.2f}")
```
