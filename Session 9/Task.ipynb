```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.preprocessing import StandardScaler
```

```python
# TASK 1: Load and Split Dataset
df = pd.read_csv('zomato_restaurant_data.csv')

# Separate into features (X) and target (y)
X = df.drop(columns=['Restaurant_Rating'])
y = df['Restaurant_Rating']

# CRITICAL STEP: Regularization requires feature scaling because large scale
# variables are unfairly penalized compared to low scale variables.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
feature_names = X.columns

# Convert scaled arrays back to dataframe for easier visualization down the line
X_scaled_df = pd.DataFrame(X_scaled, columns=feature_names)
```

```python
# TASK 2, 3, & 4: Fit Lasso, Ridge, ElasticNet & Compare Coefficients
# We use a controlled regularization strength (alpha=0.05) to clearly highlight behavior
alpha_val = 0.05

# 1. Lasso Regression (L1 Penalty)
lasso = Lasso(alpha=alpha_val, random_state=42)
lasso.fit(X_scaled, y)

# 2. Ridge Regression (L2 Penalty)
ridge = Ridge(alpha=alpha_val)
ridge.fit(X_scaled, y)

# 3. ElasticNet Regression (Combined L1 + L2)
# l1_ratio=0.5 splits penalty 50% Lasso and 50% Ridge
elastic = ElasticNet(alpha=alpha_val, l1_ratio=0.5, random_state=42)
elastic.fit(X_scaled, y)

# Combine coefficients into a clear comparative dataframe
coefficient_comparison = pd.DataFrame({
    'Feature': feature_names,
    'Lasso (L1)': lasso.coef_,
    'Ridge (L2)': ridge.coef_,
    'ElasticNet': elastic.coef_
})

print("=== TASKS 2, 3 & 4: COEFFICIENT SHRINKAGE COMPARISON ===")
print(coefficient_comparison.round(4).to_string(index=False))
print("\n" + "="*60 + "\n")
```

```python
# TASK 5: Feature Selector Using Lasso

kept_features = []
dropped_features = []

for feat, coef in zip(feature_names, lasso.coef_):
    if coef != 0:
        kept_features.append((feat, round(coef, 4)))
    else:
        dropped_features.append(feat)

print("KEPT FEATURES (Non-zero coefficients):")
for feat, coef in kept_features:
    print(f"  - {feat:<25} (Weight Impact: {coef})")

print("\nDROPPED FEATURES (Coefficients set exactly to 0 by Lasso):")
for feat in dropped_features:
    print(f"  - {feat}")
```
