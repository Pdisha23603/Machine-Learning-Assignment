```python
#Task 1: Iris Dataset using KNeighborsClassifier (5-Fold CV)

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

# Load Dataset
iris = load_iris()

X = iris.data
y = iris.target

# Model
knn = KNeighborsClassifier()

# 5-Fold Cross Validation
scores = cross_val_score(
    knn,
    X,
    y,
    cv=5
)

print("Scores:", scores)
print("Average Accuracy:", scores.mean())
```

```python
#Task 2: DecisionTreeClassifier with 10-Fold CV
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# Dataset
wine = load_wine()

X = wine.data
y = wine.target

# Model
dt = DecisionTreeClassifier(random_state=42)

# 10-Fold CV
scores = cross_val_score(
    dt,
    X,
    y,
    cv=10
)

print("Accuracy for each fold:")

for i, score in enumerate(scores, start=1):
    print(f"Fold {i}: {score:.4f}")

print("\nAverage Accuracy:", scores.mean())
```

```python
#Task 3: StratifiedKFold on Iris Dataset
from sklearn.datasets import load_iris
from sklearn.model_selection import StratifiedKFold
import numpy as np

iris = load_iris()

X = iris.data
y = iris.target

skf = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), start=1):

    y_test = y[test_idx]

    unique, counts = np.unique(
        y_test,
        return_counts=True
    )

    print(f"\nFold {fold}")

    for cls, count in zip(unique, counts):
        print(
            f"Class {cls}: {count}"
        )
```

```python
#Task 4: Compare cv = 3, 5, 10 using RandomForestClassifier

from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

wine = load_wine()

X = wine.data
y = wine.target

rf = RandomForestClassifier(
    random_state=42
)

cv_values = [3, 5, 10]

for cv in cv_values:

    scores = cross_val_score(
        rf,
        X,
        y,
        cv=cv
    )

    print("\nCV =", cv)

    print("Scores:", scores)

    print(
        "Mean Accuracy:",
        round(scores.mean(),4)
    )

    print(
        "Standard Deviation:",
        round(scores.std(),4)
    )
```

```python
#Task 5: Generic Cross Validation Function
from sklearn.model_selection import cross_val_score
import numpy as np

def evaluate_model(
    model,
    X,
    y,
    cv=5
):
    scores = cross_val_score(
        model,
        X,
        y,
        cv=cv
    )

    mean_score = np.mean(scores)

    std_score = np.std(scores)

    return mean_score, std_score


# Example Usage

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()

X = iris.data
y = iris.target

model = KNeighborsClassifier()

mean_acc, std_acc = evaluate_model(
    model,
    X,
    y,
    cv=5
)

print("Mean Accuracy:", mean_acc)

print("Standard Deviation:", std_acc)
```
